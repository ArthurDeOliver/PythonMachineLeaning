```python
#Bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
```

```python
#abertura do arquivo risco crédito 
import pickle
with open('risco_credito.pkl','rb') as f:
    X_risco_credito, y_risco_credito = pickle.load(f)
```

```python
#veriicação das variáveis
X_risco_credito, y_risco_credito
```

```python
# Criando a árvore de decisão usando o critério da entropia
arvore_risco_credito = DecisionTreeClassifier(criterion='entropy')
arvore_risco_credito.fit(X_risco_credito, y_risco_credito)
```

```python
#importância dos atributos
arvore_risco_credito.feature_importances_
```

```python
# Visualização da árvore 
from sklearn import tree

previsores = ['hisotoria', 'divida','garantias','renda'] #nome dos atributos na árvore

figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(10,10))# aumentando a visualização da árvore

tree.plot_tree(arvore_risco_credito, feature_names=previsores, class_names= arvore_risco_credito.classes_, filled=True);
#plotação da árvore (a própria árvore de decisão, nome dos atributos definidos, nome das classes)
```

```python
#colocando dados para previsões 
previsoes = arvore_risco_credito.predict([[0,0,1,2],[2,0,0,0]])
previsoes
```
